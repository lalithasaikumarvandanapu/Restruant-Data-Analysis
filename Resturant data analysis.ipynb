# When does people pay more tip i.e during lunch or dinner

```python
tips.groupby(['time'])['tip'].mean()
```

You've calculated the mean (average) value of the "tip" column for each unique value in the "time" column, using a groupby operation on the "tips" dataset. This provides the average tip for each time category.

```python
tips.groupby(['time'])['tip'].median()
```

You've calculated the median value of the "tip" column for each unique value in the "time" column, using a groupby operation on the "tips" dataset. This gives the middle value of the tips for each time category, which can be more robust to outliers than the mean.

# Lets plot these values

```python
tips.groupby(['time'])['tip'].mean().plot(kind='bar')
```

You've created a bar plot of the mean (average) values of the "tip" column for each unique value in the "time" column. This visual representation helps to compare the average tips between different time categories.

# Lets create a cross table for smoker and sex for average tips

```python
tips.groupby(['smoker','sex'])['tip'].mean()
```

You've calculated the mean (average) value of the "tip" column for each combination of unique values in the "smoker" and "sex" columns, using a groupby operation on the "tips" dataset. This provides the average tip for each combination of smoking status and gender.

```python
tips.groupby(['day','time'])['tip'].mean()
```

You've calculated the mean (average) value of the "tip" column for each combination of unique values in the "day" and "time" columns, using a groupby operation on the "tips" dataset. This provides the average tip for each combination of day and time.

# Lets put this information in pivot structure

```python
tips.groupby(['time','sex'])['tip'].mean().unstack()
```

You've performed a groupby operation on the "tips" dataset, grouping by the "time" and "sex" columns, and then calculated the mean (average) value of the "tip" column for each combination. The `unstack()` method is used to pivot the resulting Series into a DataFrame for better readability and comparison between time categories and genders.

```python
tips.groupby(['smoker','sex'])['tip'].mean().unstack()
```

# Visualization using seaborn lib

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

You've created a scatter plot using Seaborn's `scatterplot` function with "total_bill" on the x-axis and "tip" on the y-axis from the "tips" dataset. This is another way to visualize the relationship between the total bill and tip amount, leveraging the features of the Seaborn library for enhanced aesthetics and additional functionalities.

# Lets add the effect of column sex to understand data better

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
```

You've extended the scatter plot by adding the 'hue' parameter, which differentiates data points by the 'sex' variable. This creates a scatter plot where points are colored based on the gender of the individuals in the "tips" dataset, providing additional information about the distribution of total bill and tip amount with respect to gender.

# Similarly lets see visualization by Time

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='time')
```

You've created a scatter plot using Seaborn's `scatterplot` function with "total_bill" on the x-axis, "tip" on the y-axis, and points colored by the 'time' variable from the "tips" dataset. This allows you to explore the relationship between total bill and tip amount while differentiating points based on the time of day (e.g., lunch or dinner).

# Create the bifurication of time on the chart above column based on sex

```python
sns.relplot( x="total_bill",y="tip",data=tips,col="sex",hue="time")
```

You've used Seaborn's `relplot` function to create a scatter plot grid. Each subplot corresponds to a different level of the "sex" column, and the points are colored based on the 'time' variable. This allows for a more detailed exploration of the relationship between "total_bill" and "tip" for different genders and times.

# Lets look at the distribution of total bill by time

The `col` parameter is used to create separate columns for different levels of the 'time' variable. The `kde=True` indicates that kernel density estimates should be overlaid on the histograms for a smoother representation of the distribution.
